```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Gambar_Diri.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
tinggi, lebar = img_rgb.shape[:2]

plt.imshow(img_rgb)
plt.title('Citra Asli Asli')
plt.axis('off')
plt.show()
```

```python
pusat = (lebar // 2, tinggi // 2)
matriks_rotasi = cv2.getRotationMatrix2D(pusat, -45, 1.0)
img_rotasi = cv2.warpAffine(img_rgb, matriks_rotasi, (lebar, tinggi))

plt.imshow(img_rotasi)
plt.title('After Rotate')
plt.axis('off')
plt.show()
```

```python
img_resize = cv2.resize(img_rgb, (lebar // 2, tinggi // 2))

plt.imshow(img_resize)
plt.title('After Resize (50%)')
plt.axis('off')
plt.show()
```

```python
img_crop = img_rgb[tinggi//4:3*tinggi//4, lebar//4:3*lebar//4]

plt.imshow(img_crop)
plt.title('After Cropped (Tengah)')
plt.axis('off')
plt.show()
```

```python
img_flip = cv2.flip(img_rgb, 1)

plt.imshow(img_flip)
plt.title('After Flipped (Cermin)')
plt.axis('off')
plt.show()
```

```python
matriks_geser = np.float32([[1, 0, 100], [0, 1, 50]])
img_geser = cv2.warpAffine(img_rgb, matriks_geser, (lebar, tinggi))

plt.imshow(img_geser)
plt.title('After Translate')
plt.axis('off')
plt.show()
```
